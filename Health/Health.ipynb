```python
df['Health Indicator'] = df['Health Indicator'].fillna('Unknown')            
df['Holding_Policy_Duration'] = df['Holding_Policy_Duration'].fillna('Unknown')
df['Holding_Policy_Type'] = df['Holding_Policy_Type'].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,4))
sns.countplot(x='Response', data=df)
plt.title('Distribution of Response (Target)')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Accomodation_Type', hue='Response',data=df)
plt.title('Response by Accomodation Type')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Reco_Insurance_Type', hue='Response',data=df)
plt.title('Response by Reco Insurance Type')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Reco_Policy_Premium'],bins=30, kde=True)
plt.title('Distribution of Policy Premium')
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.boxplot(x='Response',y='Reco_Policy_Premium',data=df)
plt.title('Policy Premium by Response')
plt.show()
```
